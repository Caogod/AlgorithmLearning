```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA,FastICA
from mpl_toolkits.mplot3d import Axes3D
```

```python
G=[[np.cos(0.1*i) for i in range(20,3600+20,5)],
   [np.sin(0.02*i) for i in range(60,3600+60,5)],
   [np.sign(np.sin(0.005*i)) for i in range(80,3600+80,5)]]

G = np.array(G).T
A = G + 0.2 * np.random.normal(size=G.shape)  # Add noise
W = np.random.randint(1,10,(3,3))
X = np.dot(A,W)
```

```python
for i in range(G.shape[1]):
    plt.subplot(3,1,i+1)
    plt.plot(G[:,i].T)
```

```python
for i in range(A.shape[1]):
    plt.subplot(3,1,i+1)
    plt.plot(A[:,i].T)
```

```python
for i in range(X.shape[1]):
    plt.subplot(3,1,i+1)
    plt.plot(X[:,i].T)
```

```python
for i in range(X.shape[1]):
    plt.plot(X[:,i].T)
```

```python
ica = FastICA(n_components=3)
```

```python
S_ = ica.fit_transform(X)  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix
```

```python
S_.shape
```

```python
#检查原矩阵与分解矩阵是否一致
assert np.allclose(X, np.dot(S_, A_.T) + ica.mean_)
```

```python
#S为还原后的矩阵
S = np.dot(S_, A_.T)+ica.mean_
```

```python
np.allclose(S,X)
```
